```python
import ipywidgets as widgets
from IPython.display import display
from ipywidgets import interact, interactive, fixed, interact_manual

ingredients = ["aa",'b','c', 'd']

w = widgets.Dropdown(
    options=ingredients,
    description='ingredient 1',
    disabled=False
)

w2 = widgets.Dropdown(
    options= ingredients,
    description='ingredient 2',
    disabled=False)


button = widgets.Button(
    description='Click me',
    disabled=False,
    button_style='', # 'success', 'info', 'warning', 'danger' or ''
    tooltip='Click me'
)

res = widgets.HTML(
    value="Hello <b>World</b>",
)

file = open("images/us.png", "rb")
image = file.read()
file = open("images/fr.png", "rb")
image2 = file.read()
res_img = widgets.Image(
    value=image,
    format='png'
)


display(w)
display(w2)
def on_button_clicked(b):
    res.value = '<b>' + w.label + '</b>'
    res_img.value=image2

button.on_click(on_button_clicked)
display(button)
display(res)
display(res_img)
```
